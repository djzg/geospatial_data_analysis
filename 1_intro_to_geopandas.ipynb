GeoDataFrame contains a tabular, geospatial dataset:
- it has a 'geometry' column that holds the geometry information
- other columns are the attributes that describe each of the geometries

It is similar to pandas DataFrame but with some additional functionality for working with geospatial data:
- A .geometry attribute that always returns the column with the geometry information
- Extra methods for working with spatial data (area, distance, buffer, intersection,...)

### Shapely library

```python
p = Point(1, 1)
polygon = Polygon([(1, 1), (2, 2), (2, 1)])
print(p)
print(polygon)
```

### Coordinate reference systems (CRS)

Determines how the 2D coordinates of the geometry objects should be related to actual places on the Earth.

```python
countries.crs
```

```python
countries = countries[countries['NAME'] != 'Antarctica']
```

```python
# changing projection
countries_mercator = countries.to_crs(epsg=3395)
countries_mercator.plot()
```

### Plotting different layers together

```python
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15,10))
rivers.plot(ax=ax)
cities.plot(ax=ax, color='red')
ax.set(xlim=(-20, 60), ylim=(-40, 40))
```
